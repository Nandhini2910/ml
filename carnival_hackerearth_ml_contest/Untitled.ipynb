```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv("./train.csv")
test=pd.read_csv("./test.csv")
```

```python
df1=df.drop(["Product_id","instock_date","Customer_name"],axis="columns")
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df1.Loyalty_customer=le.fit_transform(df1.Loyalty_customer)
```

```python
df1.head()
```

```python
dummies=pd.get_dummies(df1.Product_Category)
df2=pd.concat([df1,dummies],axis='columns')
df3=df2.drop(['Product_Category','Technology'],axis='columns')
```

```python
df3.head()
```

```python
df3.isnull().sum()
```

```python
df3.Stall_no           = df3.Stall_no.fillna(df3.Stall_no.mean())
df3.Discount_avail     = df3.Discount_avail.fillna(df3.Discount_avail.mean())
df3.charges_1          = df3.charges_1.fillna(df3.charges_1.mean())
df3.charges_2          = df3.charges_2.fillna(df3.charges_2.mean())
df3.Minimum_price      = df3.Minimum_price.fillna(df3.Minimum_price.mean())
df3.Maximum_price      = df3.Maximum_price.fillna(df3.Maximum_price.mean())
df3.Selling_Price      = df3.Selling_Price.fillna(df3.Selling_Price.mean())
```

```python
df3.isnull().sum()
```

```python
df3.head()
```

```python
df4=df3.copy()
df4=df4.drop("Selling_Price",axis="columns")
X=df4.iloc[:,:].values
y=df3[["Selling_Price"]].values
```

```python
from sklearn.preprocessing import MinMaxScaler
x_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()
X=x_scaler.fit_transform(X)
y=y_scaler.fit_transform(y)
```

```python
model = keras.Sequential([
    keras.layers.Dense(200, input_shape=(19,), activation='relu'),
    keras.layers.Dense(200, input_shape=(200,), activation='relu'),
    keras.layers.Dense(200, input_shape=(200,), activation='relu'),
    keras.layers.Dense(1,activation='relu')
])

model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mse'])

model.fit(X, y, epochs=500)
```

```python
test1=test.drop(["Product_id","instock_date","Customer_name"],axis="columns")
```

```python
test1.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

test1.Loyalty_customer=le.fit_transform(test1.Loyalty_customer)
```

```python
dummies=pd.get_dummies(test1.Product_Category)
test1=pd.concat([test1,dummies],axis='columns')
test1=test1.drop(['Product_Category','Technology'],axis='columns')
```

```python
test1.head()
```

```python
test1.Stall_no           = test1.Stall_no.fillna(test1.Stall_no.mean())
test1.Discount_avail     = test1.Discount_avail.fillna(test1.Discount_avail.mean())
test1.charges_1          = test1.charges_1.fillna(test1.charges_1.mean())
test1.charges_2          = test1.charges_2.fillna(test1.charges_2.mean())
test1.Minimum_price      = test1.Minimum_price.fillna(test1.Minimum_price.mean())
test1.Maximum_price      = test1.Maximum_price.fillna(test1.Maximum_price.mean())
```

```python
test1.isnull().sum()
```

```python
test2=test1.iloc[:,:].values
test2=x_scaler.fit_transform(test2)
```

```python
y_pred=model.predict(test2)
y_pred.shape
```

```python
y_predicted=y_scaler.inverse_transform(y_pred)
```

```python
y_predicted
```

```python
test["Selling_Price"]=y_predicted
test.head()
```
